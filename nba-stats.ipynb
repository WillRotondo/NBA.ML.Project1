```python
import pandas as pd
import numpy as np
```

```python
bball_df = pd.read_csv('all_seasons.csv')
```

```python
bball_df.head()
```

```python
#Check for missing values 
bball_df.isna().sum()
```

```python
bball_df.describe()
```

```python
# We are going to look at the average height and weight of NBA draft classes year-to-year

#For simplicity, we will eliminate Undrafted players (unclear if data set contains starting year of undrafted player)
bball_df.replace('Undrafted', np.nan, inplace=True)


bball_df.dropna( inplace=True, how='any')
bball_df.isna().sum()
```

```python
bball_df.info()
```

```python
#Group players by draft year 
GroupedData = bball_df.groupby('draft_year').agg({'player_height': 'mean', 'player_weight':'mean'}).reset_index()
```

```python
GroupedData.head()
```

```python
import matplotlib.pyplot as plt

#Let's visualize the data 
# Plot player height
plt.figure(figsize=(10, 6))
plt.scatter(GroupedData['draft_year'], GroupedData['player_height'], label='Player Height', marker='o')
plt.title('NBA Player Height Over the Years')
plt.xlabel('Draft Year')
plt.ylabel('Player Height (cm)')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(GroupedData['draft_year'], GroupedData['player_weight'], label='Player Weight', marker='x')
plt.title('NBA Player Weight Over the Years')
plt.xlabel('Draft Year')
plt.ylabel('Player Weight (kg)')
plt.legend()
plt.xticks(rotation=45)
plt.grid()
```

```python
#Let's check the outliers for 1963, 1976, 1978, 1979, and 1980-- most likely only have a few players in each year 
#Actually, it would be easier to just check the number of players in each draft class 
players_per_draft = bball_df['draft_year'].unique()
player_draft_count = {}

for year in players_per_draft: 
    players = len(bball_df[bball_df['draft_year'] == year])
    player_draft_count[year] = players

for year, count in player_draft_count.items():
    print(f'Number of players drafted in {year}: {count}')
```

```python
#Looking at this, and using an arbitrary eye test (this data is not official/up to date), we will only use the data from 1990 to 2012
GroupedData['draft_year'] = GroupedData['draft_year'].astype(int)
updated_player_group = GroupedData[(GroupedData['draft_year'] >= 1990) & (GroupedData['draft_year'] <= 2012)]

updated_player_group.info()
```

```python
#Let's visualize the data 
# Plot player height
plt.figure(figsize=(12, 6))
plt.plot(updated_player_group['draft_year'], updated_player_group['player_height'], label='Player Height')
plt.title('NBA Player Height Over the Years')
plt.xlabel('Draft Year')
plt.ylabel('Player Height (cm)')
plt.ylim(170,220)
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(updated_player_group['draft_year'], updated_player_group['player_weight'], label='Player Weight')
plt.title('NBA Player Weight Over the Years')
plt.xlabel('Draft Year')
plt.ylabel('Player Weight (kgs)')
plt.ylim(70,120)
plt.legend()
plt.xticks(rotation=45)
plt.grid()
```

```python
#Now that we have some relatively good looking data, let's try and estimate weight and height over the next few years using linear regression! 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

#Forgot to seperate the data into two data frames for weight and height 
weight_df = updated_player_group[['draft_year', 'player_weight']]
height_df = updated_player_group[['draft_year', 'player_height']]

X_weight = weight_df[['draft_year']].values
y_weight = weight_df['player_weight'].values
X_height = height_df[['draft_year']].values
y_height = height_df['player_height'].values

lin_reg = LinearRegression()

# Using cross validation, we will check if linear regression is the proper way to go about this for weight...
scores_weight = cross_val_score(lin_reg, X_weight, y_weight, cv=10, scoring='r2')

#And then for height 
scores_height = cross_val_score(lin_reg, X_height, y_height, cv=10, scoring='r2')

# Print R2 scores 
print(f'Linear Regression R2 Score for weight: {scores_weight.mean()}')
print(f'Linear Regression R2 Score for height: {scores_height.mean()}')
```

```python
# We can see that Linear Regression did not work very well... let's try some other methods! 
from sklearn.ensemble import RandomForestRegressor

random_forest = RandomForestRegressor()

# Using cross validation, we will check if linear regression is the proper way to go about this for weight...
forest_scores_weight = cross_val_score(random_forest, X_weight, y_weight, cv=10, scoring='r2')

#And then for height 
forest_scores_height = cross_val_score(random_forest, X_height, y_height, cv=10, scoring='r2')

# Print R2 scores 
print(f'Random Forest R2 Score for weight: {forest_scores_weight.mean()}')
print(f'Random Forest R2 Score for height: {forest_scores_height.mean()}')
```

```python
# Well, that didn't work very well either 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

def polynomial_regression_crossval(X, y, degree):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    r2_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
    return r2_scores.mean()

#I got better numbers for a degree 2 polynomial, but let's try and make the model more complex to see if we can get a positive R2 score 

degrees = [2,3,4,5]

for degree in degrees:
    weight_poly = polynomial_regression_crossval(X_weight, y_weight, degree)
    height_poly = polynomial_regression_crossval(X_height, y_height, degree)
    print(f'Polynomial Regression (Degree {degree}) R2 Score for weight: {weight_poly}')
    print(f'Polynomial Regression (Degree {degree}) R2 Score for height: {height_poly}')
```

```python
#Although degree 2 got a close-to-zero average R2 score, we still want to do better. Let's try KNN

from sklearn.neighbors import KNeighborsRegressor

def knn_regression_crossval(X,y,k):
    model = KNeighborsRegressor(n_neighbors=k)
    r2_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
    return r2_scores.mean()

k_values = [3, 5, 7, 9, 10]  

for k in k_values:
    weight_r2 = knn_regression_crossval(X_weight, y_weight, k)
    height_r2 = knn_regression_crossval(X_height, y_height, k)
    print(f"KNN (K={k}) - Weight R2 Score: {weight_r2}")
    print(f"KNN (K={k}) - Height R2 Score: {height_r2}")
```

```python
# Oof! It looks like our best bet is Degree 2 Polynomial Regression! Let's see how it predicts the next 10 years 

poly_model_weight = PolynomialFeatures(2)
X_weight_poly = poly_model_weight.fit_transform(X_weight)

poly_model_height = PolynomialFeatures(2)
X_height_poly = poly_model_height.fit_transform(X_height)

model_weight = LinearRegression()
model_height = LinearRegression()

# Fit 
model_weight.fit(X_weight_poly, y_weight)
model_height.fit(X_height_poly, y_height)

future_year = pd.DataFrame({'draft_year': range(2013, 2023)})

# Create polynomial features for predicted years
future_year_poly_weight = poly_model_weight.transform(np.array(future_year).reshape(-1, 1))
future_year_poly_weight = poly_model_height.transform(np.array(future_year).reshape(-1, 1))

# Predict
future_weight_predictions = model_weight.predict(future_year_poly_weight)
future_height_predictions = model_height.predict(future_year_poly_weight)

# Create DataFrames
weight_predictions_df = pd.DataFrame({'draft_year': future_year['draft_year'], 'predicted_weight': future_weight_predictions})
height_predictions_df = pd.DataFrame({'draft_year': future_year['draft_year'], 'predicted_height': future_height_predictions})

print("Weight Predictions:")
print(weight_predictions_df)

print("Height Predictions:")
print(height_predictions_df)
```

```python
# We've got our predictions! Now lets visualize it alongside the original data 

# Plot the OG data
plt.figure(figsize=(12, 6))
plt.plot(X_weight, y_weight, label='Original Weight Data', marker='o')
plt.plot(X_height, y_height, label='Original Height Data', marker='o')

# Plot the predictions
plt.plot(weight_predictions_df['draft_year'], weight_predictions_df['predicted_weight'], label='Predicted Weight(kg)', marker='x')
plt.plot(height_predictions_df['draft_year'], height_predictions_df['predicted_height'], label='Predicted Height(cm)', marker='x')

plt.xlabel('Draft Year')
plt.ylabel('Player Weight and Height')
plt.title('Player Weight and Height Predictions vs. Actual Data')
plt.legend()
plt.grid(True)

plt.show()
```
